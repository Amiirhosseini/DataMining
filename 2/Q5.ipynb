```python
import pandas as pd
```

1-read the csv file and store it in a dataframe and show the Data frame

```python
titanic = pd.read_csv('./titanic.csv', sep=',')

titanic.head()
```

2- delete rows with null values even if only one column has null value
Also remove 'Cabin' column from data table<br/>
PreProcessing:

```python
titanic = titanic.drop(['Cabin'], axis=1)
titanic
```

```python
titanic.isnull().sum()
```

```python
titanic = titanic.dropna()

titanic
```

```python
titanic.isnull().sum()
```

3- remove the 'Parch', 'Sibsp', 'Ticket', 'Name'  and 'Passengerid' columns from the titanic dataframe

```python
titanic = titanic.drop(['Parch', 'SibSp', 'Ticket', 'Name', 'PassengerId'], axis=1)
titanic
```

4- scale categorical data with one hot encoding in 2 ways

```python
# from sklearn.preprocessing import OneHotEncoder
# from sklearn.preprocessing import LabelEncoder

# #convert categorical data to numerical data in dataframe
# #convert 'sex','Cabin','Embarked' column to numerical data

# label_encoder = LabelEncoder()
# titanic['Sex'] = label_encoder.fit_transform(titanic['Sex'])
# titanic['Cabin'] = label_encoder.fit_transform(titanic['Cabin'])
# titanic['Embarked'] = label_encoder.fit_transform(titanic['Embarked'])

# enc = OneHotEncoder(handle_unknown='ignore')
# titanic['Sex'] = enc.fit_transform(titanic['Sex'].values.reshape(len(titanic['Sex']),1)).toarray()
# titanic['Cabin'] = enc.fit_transform(titanic['Cabin'].values.reshape(len(titanic['Cabin']),1)).toarray()
# titanic['Embarked'] = enc.fit_transform(titanic['Embarked'].values.reshape(len(titanic['Embarked']),1)).toarray()
# titanic
```

```python
titanic = pd.get_dummies(titanic, columns=['Sex','Embarked'])

titanic
```

5- Categorized every columns to x except the 'Survived' column

```python
Y = titanic['Survived']
X = titanic.drop(['Survived'],axis=1)
```

6- Split the data in 30% of test and 70% of train

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.3)

print("The shape of X_train is:", X_train.shape)
print("The shape of X_test is:", X_test.shape)
print("The shape of Y_train is:", Y_train.shape)
print("The shape of Y_test is:", Y_test.shape)
```

7- Create desition tree model with random_state=0 <br>
calculate accuracy of model for testing and training data

```python
from sklearn import tree
model = tree.DecisionTreeClassifier(random_state=0)

#train model with training data
model.fit(X_train,Y_train)
#calculate accuracy of model for training data
from sklearn.metrics import accuracy_score, precision_score
Y_pred_train = model.predict(X_train)
print("Accuracy of model for training data: ",accuracy_score(Y_train,Y_pred_train))
print("Precision of model for training data: ",precision_score(Y_train,Y_pred_train))

#calculate accuracy of model for testing data
Y_pred_test = model.predict(X_test)
print("Accuracy of model for testing data: ",accuracy_score(Y_test,Y_pred_test))
print("Precision of model for testing data: ",precision_score(Y_test,Y_pred_test))
```

8-  plotting the tree with graphViz and pydotplus (needs installing executable binary files for graphviz)

```python
import pydotplus
from IPython.display import Image

dot_data = tree.export_graphviz(model, out_file=None, feature_names=X.columns, class_names=['0','1'], filled=True, rounded=True, special_characters=True)
graph = pydotplus.graph_from_dot_data(dot_data)
graph.write_jpeg('Q5.jpeg')
Image(graph.create_png())
```

9- Overfitting happened:  when the model is too complex, the training error is small but the test error is large <br/>
As it is clear, the accuracy of the model for training data is much bigger than the accuracy of the model for testing data same for Precision <br/>
So, we can assume overfitting happened in this case.

For preventing this to happen, we can increase the size of training data or Regularization of existing data. <br/>
We can also use Early stopping methods or Pruning to keep the model fitted.

10- show confusion matrix for testing data and illustrate it

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

def plot_confusion_matrix(y,y_predict):
    cm = confusion_matrix(y, y_predict)
    ax= plt.subplot()
    sns.heatmap(cm, cmap='coolwarm',annot=True,fmt = " ", ax = ax);
    ax.set_xlabel('Predicted labels')
    ax.set_ylabel('True labels')
    ax.set_title('Confusion Matrix');
    ax.xaxis.set_ticklabels(['Predicted Died','Predicted Survived'])
```

```python
confusionMatrix = pd.DataFrame(confusion_matrix(Y_test, Y_pred_test))

confusionMatrix.index = ['True[0]','True[1]']

confusionMatrix.columns = ['Predict[0]','Predict[1]']

confusionMatrix
```

```python
plot_confusion_matrix(Y_test,Y_pred_test)
```

```python
print(classification_report(Y_pred_test, Y_test))
```

This graph shows the performance of the model over True positive and True negative labels of model. <br/>
As we can see, 113+51 of samples are True predicted and 26+23 of samples are predicted in a wrong label 

11- hyper parameters tunning : find best max_depth and min_samples_leaf by grid search

```python
from sklearn.model_selection import GridSearchCV
import numpy as np

param_grid = {'max_depth': np.arange(3, 10),'min_samples_leaf': np.arange(2, 10)}
grid = GridSearchCV(tree.DecisionTreeClassifier(random_state=0), param_grid, cv=5)
grid.fit(X_train, Y_train)
print("Best cross-validation score: {:.2f}".format(grid.best_score_))
print("Best parameters: ", grid.best_params_)
print("Best estimator: ", grid.best_estimator_)
print("Best test-set score: {:.2f}".format(grid.score(X_test, Y_test)))
```

```python
y_pred_train = grid.predict(X_train)

print("Accuracy of model for training data: ",accuracy_score(Y_train,y_pred_train))
print("Precision of model for training data: ",precision_score(Y_train,y_pred_train))

y_pred_test = grid.predict(X_test)

print("Accuracy of model for testing data: ",accuracy_score(Y_test,y_pred_test))
print("Precision of model for testing data: ",precision_score(Y_test,y_pred_test))
```

12- showing the best model gained from grid search by GraphViz

```python
best_model = grid.best_estimator_ 

dot_data = tree.export_graphviz(best_model, out_file=None, feature_names=X.columns, class_names=['0','1'], filled=True, rounded=True, special_characters=True)
graph = pydotplus.graph_from_dot_data(dot_data)
graph.write_jpeg('Q5_grid_best.jpeg')
Image(graph.create_png())
```

15-  Hyperparametes tuning : Randomized Search for RandomForest parametes

```python
from sklearn.model_selection import RandomizedSearchCV
import numpy as np

param_random = { 'max_depth':range(2,10),
    'min_samples_split':range(2,11),
    'min_samples_leaf' : range(1,11),
    'bootstrap': [True, False],
    'n_estimators' : range(1,1000),
    'criterion':['gini','entropy']}
random = RandomizedSearchCV(
    RandomForestClassifier(random_state=0), param_random,n_iter=5, n_jobs=-1)

random.fit(X_train, Y_train)
print("Best cross-validation score: {:.2f}".format(random.best_score_))
print("Best parameters: ", random.best_params_)
print("Best estimator: ", random.best_estimator_)
print("Best test-set score: {:.2f}".format(random.score(X_test, Y_test)))
```

```python
print("Accuracy on training set: {:.3f}".format(random.score(X_train, Y_train)))
print("Precision on training set: {:.3f}".format(precision_score(Y_train, random.predict(X_train))))

#precision score
print("Accuracy on test set: {:.3f}".format(random.score(X_test, Y_test)))
print("Precision on test set: {:.3f}".format(precision_score(Y_test, random.predict(X_test))))
```
