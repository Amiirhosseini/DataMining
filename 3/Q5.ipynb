```python
import pandas as pd
```

1-read the csv file and store it in a dataframe and show the Data frame info

```python
heart = pd.read_csv('./heart_diagnose.csv')
heart.head()
```

```python
heart.info()
```

```python
heart.describe()
```

```python
heart.isnull().count()
```

2- scale categorical data with one hot encoding

```python
#find categorical variables
categorical = [var for var in heart.columns if heart[var].dtype=='O']
heart = pd.get_dummies(heart,columns = categorical, drop_first=True)
```

```python
heart
```

3- reduce features to 2 with PCA method

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
principalComponents = pca.fit_transform(heart)
new_heart = pd.DataFrame(data = principalComponents, columns = ['new f1', 'new f2'])
new_heart
```

4- find the best amount of clusters based on the elbow method

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
wcss = []
for i in range(1, 20):
    kmeans = KMeans(n_clusters = i)
    kmeans.fit(new_heart)
    wcss.append(kmeans.inertia_)
    
plt.plot(range(1, 20), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('performance')
plt.show()
```

```python
#print the best number of clusters
print("The best number of clusters is:", 3)
```

```python
#print the best number of inertia
print("The best number of inertia score is:", 100)
```

5- bar plot for silhouette score for K in range of 2-20 and find the best K

```python
from sklearn.metrics import silhouette_score
import seaborn as sns
sil = []
for k in range(2, 20):
    kmeans = KMeans(n_clusters = k).fit(new_heart)
    labels = kmeans.labels_
    sil.append(silhouette_score(new_heart, labels, metric = 'euclidean'))
plt.bar(range(2, 20), sil)
plt.title('Silhouette Method')
plt.xlabel('Number of clusters')
plt.ylabel('performance')
plt.show()
```

```python
#print the best number of clusters
print("The best number of clusters is:", 2)
```

```python
#print the best number of silhouette score
print("The best number of silhouette score is:", 0.5)
```

6- run k-means with the best number of clusters K=3,K=2 and plot the results

```python
kmeans = KMeans(n_clusters = 3)
kmeans.fit(new_heart)
y_kmeans = kmeans.predict(new_heart)
plt.scatter(new_heart['new f1'], new_heart['new f2'], c=y_kmeans, s=50, cmap='viridis')
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.5)
plt.title('K-means clustering(K=3)')
plt.xlabel('new f1')
plt.ylabel('new f2')
plt.show()
```

```python
#do the same for K=2
kmeans = KMeans(n_clusters = 2)
kmeans.fit(new_heart)
y_kmeans = kmeans.predict(new_heart)
plt.scatter(new_heart['new f1'], new_heart['new f2'], c=y_kmeans, s=50, cmap='viridis')
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.5)
plt.title('K-means clustering(K=2)')
plt.xlabel('new f1')
plt.ylabel('new f2')
plt.show()
```
