6-show the heatmap chart of attributes and find strong relations

```python
import seaborn as sns
import matplotlib as mp
```

```python
sns.heatmap(smartphone.corr(), annot=True,linewidths=0.5, cmap='YlGnBu')
mp.pyplot.gcf().set_size_inches(15, 15)
mp.pyplot.tight_layout()
mp.pyplot.show()
```

**Report from heatmap:** <br>

Based on the above chart, it can be seen that the blue squares are somehow shows the attributes that are related together like: <br>
four_g & three_g - pc & fc -  screen height (sc_h) & screen width(sc_w) - pixel resolution height (px_height) & pixel resolution width (px_width) - ram & price_range

choose the attributes with high correlations upper than 0.5

```python
df2=smartphone[['four_g','three_g','pc','fc','sc_h','sc_w','px_height','px_width','ram','price_range']]
df2
```

```python
sns.pairplot(df2)
```

```python
from scipy.stats import spearmanr

#calculate spearman correlation for each attributes of df2
for i in df2.columns:
    for j in df2.columns:
        if i != j:
            corr, _ = spearmanr(df2[i], df2[j])
            print('Spearmans correlation between %s and %s: %.3f' % (i, j, corr))
```
