10- use other strategies to fill NaN records

strategy: str, default=’mean’

    The imputation strategy.

        If “mean”, then replace missing values using the mean along each column. Can only be used with numeric data.

        If “median”, then replace missing values using the median along each column. Can only be used with numeric data.

        If “most_frequent”, then replace missing using the most frequent value along each column. Can be used with strings or numeric data. If there is more than one such value, only the smallest is returned.

        If “constant”, then replace missing values with fill_value. Can be used with strings or numeric data.

    New in version 0.20: strategy=”constant” for fixed value imputation.


-median strategy:

```python
df_1to5_column_median = df.iloc[:,1:6]
#fill the NaN values with the mean of each column with SimpleImputer
imputer = SimpleImputer(missing_values=np.nan, strategy='median')
imputer.fit(df_1to5_column_median)
df_1to5_column_median = imputer.transform(df_1to5_column_median)
#change the array to dataframe
df_1to5_column_median = pd.DataFrame(df_1to5_column_median)
df_1to5_column_median
```

-most frequent strategy:

```python
df_1to5_column_mode = df.iloc[:,1:6]
#fill the NaN values with the mean of each column with SimpleImputer
imputer = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
imputer.fit(df_1to5_column_mode)
#change the array to dataframe
df_1to5_column_mode = imputer.transform(df_1to5_column_mode)
df_1to5_column_mode = pd.DataFrame(df_1to5_column_mode)
df_1to5_column_mode
```

-constant strategy:

```python
df_1to5_column_const = df.iloc[:,1:6]
#fill the NaN values with the mean of each column with SimpleImputer
imputer = SimpleImputer(missing_values=np.nan, strategy='constant', fill_value=5)
imputer.fit(df_1to5_column_const)
df_1to5_column_const = imputer.transform(df_1to5_column_const)
#change the array to dataframe
df_1to5_column_const = pd.DataFrame(df_1to5_column_const)
df_1to5_column_const
```

11- illustrate which one of above strategies is best for tackle with missing values

There is no one-size-fits-all answer to the question of which imputation strategy is best for tackling missing values. The choice of imputation method depends on various factors such as the type of data, the amount and pattern of missingness, the research question, and the analysis model.

For instance, mean imputation may be appropriate when the data are normally distributed and the missing values are not too extreme, while median imputation may be preferred when the data are skewed or contain outliers. Most frequent imputation may be suitable for categorical data, while constant imputation may be useful when the missing values are a result of a known mechanism or process, such as survey nonresponse.

However, it's important to note that all of these methods have limitations and potential drawbacks. Imputing missing values can introduce bias, distortions in variance, and affect statistical power. Therefore, it's advisable to explore multiple imputation methods and compare their results to understand the sensitivity of the analysis to the choice of imputation method.
